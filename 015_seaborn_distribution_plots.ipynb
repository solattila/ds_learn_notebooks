# Distribution plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DATA/dm_office_sales.csv')
df.head()
```

```python
sns.rugplot(x='salary', data=df)
```

```python
plt.figure(figsize=(8,5), dpi=100)
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.displot(data=df, x='salary', bins=20)
```

```python
sns.set(style='darkgrid')
sns.displot(data=df, x='salary', bins=20)
```

```python
sns.displot(data=df, x='salary', bins=20, color='blue', edgecolor='red', linewidth=2)
```

```python
sns.histplot(data=df, x='salary')
```

```python
sns.displot(data=df, x='salary', bins=20, kde=True)
```

```python
sns.displot(data=df, x='salary', bins=20, kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x='age')
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age')
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100])
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.1)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=1.5)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=20)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.6, shade=True)
```
