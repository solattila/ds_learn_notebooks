```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 40)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')

```

## Subplots

```python
a = np.linspace(0, 10, 11)
b = a ** 4
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, y)
axes[0, 1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, y)
axes[0, 1].plot(a, b)
axes[1, 0].plot(a, b)
axes[1, 1].plot(a, b)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, y)
axes[0, 1].plot(a, b)
axes[1, 0].plot(a, b)
axes[1, 1].plot(a, b)

fig.subplots_adjust(wspace=1, hspace=0.5)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, y)
axes[0, 0].set_ylabel('y 0 0')
axes[0, 1].plot(a, b)
axes[0, 1].set_title('Title 0 1')
axes[1, 0].plot(a, b)
axes[1, 1].plot(a, b)

fig.suptitle('Figure level title')

plt.tight_layout()

fig.savefig('saved_data/subplots_example.png', bbox_inches='tight')
```

## Legends

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')

ax.plot(x, x ** 2, label='X vs X^2')

ax.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='#650dbd', label='x vs x', lw=5, ls='--')

ax.plot(x, x+1, color='purple', label='x vs x+1', linewidth=8, linestyle='-.')

lines = ax.plot(x, x-1, linewidth=2)

lines[0].set_dashes([1,2,1,2,10,2])

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='#650dbd', label='x vs x', lw=2, marker='o', ls='--', ms=10, markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')

ax.plot(x, x + 1, color='#650dbd', label='x vs x+1', lw=2, marker='+', ls='--', markersize=20)

ax.legend()
```
