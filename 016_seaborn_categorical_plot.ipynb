# Categorical plot
## Statistics within categories

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DATA/dm_office_sales.csv')
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.countplot(data=df, x='division')
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.countplot(data=df, x='division')
plt.ylim(90, 260)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.countplot(data=df, x='level of education')
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.countplot(data=df, x='level of education', hue='division')
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd')
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')
plt.legend(bbox_to_anchor=(1.05, 1))
```

## Distribution within categories

```python
df = pd.read_csv('DATA/StudentsPerformance.csv')
df.head()
```

```python
sns.boxplot(data=df, y='math score')
```

```python
plt.figure(figsize=(10,6), dpi=100)
sns.boxplot(data=df, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(10,6), dpi=100)
sns.boxplot(data=df, y='reading score', x='parental level of education', hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,6), dpi=100)
sns.boxplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', palette='Set2')

plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,6), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='test preparation course')

plt.legend(bbox_to_anchor=(1.2,1))
```

```python
plt.figure(figsize=(10,6), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', split=True, inner='quartile')

plt.legend(bbox_to_anchor=(1.2,1))
```

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', bw=0.1)

plt.legend(bbox_to_anchor=(1.2, 1))
```

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', bw=1)

plt.legend(bbox_to_anchor=(1.2, 1))
```

```python
sns.swarmplot(data=df, x='math score')
```

```python
plt.figure(figsize=(8, 4), dpi=100)
sns.swarmplot(data=df, x='math score', y='gender', size=2)
```

```python
plt.figure(figsize=(8, 4), dpi=100)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course')

plt.legend(bbox_to_anchor=(1.3, 1))
```

```python
plt.figure(figsize=(8, 4), dpi=100)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.3, 1))
```

```python
plt.figure(figsize=(8, 4), dpi=100)
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='gender')
```
